## 2. Select 80 curves from each class

```python
random.seed(1)
col1 = sorted(random.sample([x for x in df1.columns], 80))
col2 = sorted(random.sample([x for x in df2.columns], 80))
col3 = sorted(random.sample([x for x in df3.columns], 80))
col4 = sorted(random.sample([x for x in df4.columns], 80))
col5 = sorted(random.sample([x for x in df5.columns], 80))
col6 = sorted(random.sample([x for x in df6.columns], 80))
col7 = sorted(random.sample([x for x in df7.columns], 80))
```

```python
print(len(col1), len(col2), len(col3), len(col4), len(col5), len(col6), len(col7))
```

```python
df1ae = df1[col1]
df2ae = df2[col2]
df3ae = df3[col3]
df4ae = df4[col4]
df5ae = df5[col5]
df6ae = df6[col6]
df7ae = df7[col7]
```
